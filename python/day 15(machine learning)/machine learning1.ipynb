```python
#ML:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import *
print("done")
```

```python
data={'study_hrs':range(1,11),
     'marks_obtained':[25,33,46,50,55,67,78,85,95,100]}
df=pd.DataFrame(data)
df
```

```python
plt.title('study vs marks')
plt.scatter(df['study_hrs'],df['marks_obtained'],marker='x',color='red')
plt.xlabel('study hours')
plt.ylabel('marks obtained')
plt.show()
```

```python
 #x(1),y(1):simple linear regression
#x(multiple),y(1):multiple linear regression
df
```

```python
#importing module
from sklearn.linear_model import LinearRegression
#object model creation
model=LinearRegression()
#x,y
x=df[['study_hrs']]
y=df['marks_obtained']
#training/learning
model.fit(x,y)
```

```python
m=model.coef_
c=model.intercept_
print(m,c)
```

```python
# y = mx + c
# "y = β₀ + β₁x"
# y = wx +  c
x=[[5.5]]
marks=m*x+c
print("predicted marks:",marks[0][0])
```

```python
import warnings
warnings.filterwarnings('ignore')
study_hrs = float(input('Enter study hours: '))

std_hr = [[study_hrs]]

ans = model.predict(std_hr)
marks = round(ans[0],2)

if marks > 100:
  marks = 100

elif marks<0:
  marks = 0

else:
  marks = marks

print(f'Your Predicted marks is: {marks}')
```
