```python
class Remote:
    def __init__(self):
        print("object has been created")
        print("now you can use it")
    def vol_up(self):
        print("volume up by 1 unit")
    def vol_down(self):
        print("volume down by 1 unit")
    def power_on(self):
        print("swiched on")
    def power_off(self):
        print("switched off")
    def auto_suggest_channel(self):
        import time
        print("searching trending channel")
        time.sleep(4)
        print("YOUTUBE||IITM")
    def __del__(self):
        print("object has been destroyed successfully")
```

```python
mi=Remote()
print("done")
```

```python
mi.power_on()
mi.power_off()
mi.vol_down()
mi.vol_up()
mi.auto_suggest_channel()
```

```python
del(mi)
```
